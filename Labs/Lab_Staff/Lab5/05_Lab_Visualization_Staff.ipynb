# 1 Simple Line Plots

The simpleast and commonly used engineering  plot is the good old simple line plots. 

Let's start with this simple line plot.


```python
%matplotlib inline
#import library required
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') #set t he stype to use
import numpy as np

fig=plt.figure() 
ax=plt.axes()
x=np.linspace(0.0, 10.0, num=20)
print(x)
y=2*x+5
ax.plot(x,y)
plt.show()
```

In matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container which contains all the objects representing axes, graphics, text, labels, etc. The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which will eventually contain other plot elements. Through this course, we’ll commonly use the variable name fig to refer to a figure instance, and ax to refer to an axes instance or set of axes instances.

Once we have created an axes, we can use the ax.plot function to plot some data as shown in the cell above.

# 2 Customized the Axis
Matplotlib does a fairly good job of choosing default axes limits for your plot, but sometimes it’s nice to have finer control. Here we’ll briefly see how to change the limits of the x and y axes. The most basic way to do this is to use the plt.xlim() and plt.ylim() methods to set the numerical limits of the x and y axes.



```python
ax=plt.axes()
ax.plot(x, np.sin(x)) #plot a sin wave
ax.set_xlim([-1, 11])    #set x limit to -1 to 11
ax.set_ylim([-1.5, 1.5]);  #set y limit to -1.5 to 1.5
```

```python
fig=plt.figure() 
ax=plt.axes()
ax.plot(x, np.sin(x))

ax.set_xlim(10, 0)    # if you want to reverse the order of the x axis from 10 to 0
ax.set_ylim(1.2, -1.2);  #if you want ot reverse the order of the y axis from 1.2 to -1.2
```

```python
#We also allowed to do things like automatically tighten the bounds around the current plo
fig=plt.figure() 
ax=plt.axes()
ax.plot(x, np.sin(x))
ax.axis('tight');
```

From the graph we can see that there is not much change in the mean temperature, it is nearly a straight line. We are unablet o observe any obvious relationship between the two parameters. # 2 Simple Scatter plot

We are going to work on another commonly used simple Scatter plot. Simple scatter plot is very similar with line plot, instead of points being joined by line segment, the points are respresentated individually with a dot, or othe shape. Of course these individual point can be join by line segment which is essential a line plot.

To plot scatter plots when markers are identical in size and color, we use the plot method because:

- The plot function will be faster for scatterplots where markers don't vary in size or color.
- Any or all of x, y, s, and c may be masked arrays, in which case all masks will be combined and only unmasked points will be plotted.
- Fundamentally, scatter works with 1-D arrays; x, y, s, and c may be input as 2-D arrays, but within scatter they will be flattened. The exception is c, which will be flattened only if its size matches the size of x and y.

The code below would plot a series of randomly generated data using all the symbols available in the library.


```python
fig=plt.figure() 
rng=np.random.RandomState(0)
ax=plt.axes()
for m in ['o','.',',','x','+','v','^','<','>','s','d']:
    ax.plot(rng.rand(5),rng.rand(5),m,
    label="m='{0}'".format(m))
    ax.legend(numpoints=1)
```

# 3 Histogram

### What is a histogram?

A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data. This allows the inspection of the data for its underlying distribution (e.g., normal distribution), outliers, skewness, etc.

Refor to the lnk for full documentation.

[https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html]https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html)



```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data=np.random.randn(1000)
fig=plt.figure() 
ax=plt.axes()
ax.hist(data)
ax.set_title('Histogram of Random number')
ax.set_xlabel('Bin')
ax.set_ylabel('Frequency')
```

```python
#more control over the look and feel of the histogram
#20 bin with and density =true is to have probability density=1 

fig=plt.figure() 
ax=plt.axes()
ax.hist(data,bins=15,density=True,alpha=0.5,histtype='bar',edgecolor='black',color='r')
ax.set_title('Histogram of Random number')
ax.set_xlabel('Bin')
ax.set_ylabel('Frequency')
```

# 5 Subplot

We plot line plot and histogram on two seperately in the previous sections. In this section, we see how we can create multi-pot with the same figure.

```python
#Creates two subplots and unpacks the output array immediately with sharing same X axis
f, (ax1, ax2,ax3) = plt.subplots(3, 1, sharex=True)   #3 rows 1 column
ax1.plot(gdata.index,gdata['Open'])
ax2.plot(gdata.index,gdata['Close'])
ax1.set_title('Sharing X axis')
ax3.plot(gdata.index,gdata['High'],'-o')
```

```python
#Creates two subplots and unpacks the output array immediately with sharing same Y axis
f, (ax1, ax2,ax3) = plt.subplots(1, 3, sharey=True)  #1 row 3 column
ax1.plot(gdata.index,gdata['Open'])
ax2.plot(gdata.index,gdata['Close'])
ax1.set_title('Sharing Y axis')
ax3.plot(gdata.index,gdata['High'],'-o')
```

```python
#Creates two subplots and unpacks the output array immediately with sharing same Y axis
f, (ax1, ax2,ax3) = plt.subplots(3, 1, sharey=True,sharex=True)
ax1.plot(gdata.index,gdata['Open'])
ax2.plot(gdata.index,gdata['Close'])
ax1.set_title('Sharing Y axis')
ax3.plot(gdata.index,gdata['High'],'-o')
```

```python
# Four axes, returned as a 2-d array
f, axarr = plt.subplots(2, 2)          #2x2 plots
f.set_size_inches(10, 12)
axarr[0, 0].plot(gdata.index, gdata['Open'])
axarr[0, 0].set_title('Axis [0,0]')
axarr[0, 1].plot(gdata.index, gdata['Close'])
axarr[0, 1].set_title('Axis [0,1]')
axarr[1, 0].plot(gdata.index, gdata['High'])
axarr[1, 0].set_title('Axis [1,0]')
axarr[1, 1].plot(gdata.index, gdata['Volume'])
axarr[1, 1].set_title('Axis [1,1]')
```

# 6 Bar Chart

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a line graph.

```python
import matplotlib.pyplot as plt
 
# Make a fake dataset:
product1 = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
x_pos = np.arange(len(bars))
fig, ax = plt.subplots()
# Create bars
ax.bar(x_pos, product1,label='Product1')
 
# Create names on the x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(('A', 'B', 'C','D','E'))
ax.set_title('Bar plot of Product  1 vs different parameters')
ax.set_xlabel('Parameters')
ax.set_ylabel('value')

legend = ax.legend(loc='upper center', fontsize='x-large')

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C0')
```

If there is a product2 and we want to plot bar chart of paraemter A..E.. We can plot side by side for comparision as show below:


```python
# Make a fake dataset:
product1 = [3, 12, 5, 18, 45]
product2 = [10, 5, 9, 10, 60]

bars = ('A', 'B', 'C', 'D', 'E')
width = 0.35  # the width of the bars
x_pos = np.arange(len(bars))
fig, ax = plt.subplots()
# Create bars
ax.bar(x_pos - width/2, product1,width,label='Product1',color='SkyBlue')
ax.bar(x_pos + width/2, product2,width,label='Product2',color='IndianRed')
# Create names on the x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(('A', 'B', 'C','D','E'))
ax.set_title('Bar plot of Product  1 and Product 2 vs different parameters')
ax.set_xlabel('Parameters')
ax.set_ylabel('value')
legend = ax.legend(loc='upper center',fontsize='x-large')

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C0')
```

# 7 Basic Pie Chart

A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents. While it is named for its resemblance to a pie which has been sliced, there are variations on the way it can be presented. 




```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Product1', 'Product2', 'Product3', 'Product4','product5'
sizes = [15, 30, 45, 10,20]
explode = (0, 0.1, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1 = plt.figure()
ax1=plt.axes()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
